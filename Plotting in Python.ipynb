```python
def parametric_plots(a,k):
    t = np.linspace(0,2*np.pi)
    x = 2*k*np.cos(t) - a*np.cos(k*t)
    y = 2*k*np.sin(t) - a*np.sin(k*t)
    plt.plot(x,y)
    
for N in [1,2,3,4]:
    a = np.random.uniform(2,4)
    k = np.random.uniform(2,4)
    plt.subplot(2,2,N)
    parametric_plots(a,k)

plt.axis('equal')
plt.tight_layout()
plt.show()
    
        
    
```

```python
x = np.linspace(-6,6,100)

y = np.cos(x)
y2 = 1 - x**2/2
y4 = 1-x**2/2+x**4/4

plt.plot(x,y,'r',label = 'cos(x)')
plt.plot(x,y2,label = 'taylor 2')
plt.plot(x,y4,label = 'taylor 4')
plt.grid(True,ls= ':',lw = 2)
plt.legend()
plt.xlim((-6,6))
plt.ylim((-4,4))

plt.show()
```

```python
x = np.linspace(-2,2,100)

y = np.e**(-x**2)*np.cos(2*np.pi*x)
plt.figure(figsize = (12,5))
plt.plot(x,y,c = 'r',ls = 'dashed',label = 'Decaying Cosine',marker = 'o',markerfacecolor = 'b',markersize = 5)
plt.legend()




plt.show()

```
